# User - item Matrix의 합산 문제 해결 처리

```python
main = pd.read_csv("./main_sliced.csv", index_col=0)
# 중복만 가져오기
duplicated = main[main.duplicated(['user_id','name'], keep=False)][['user_id','name','stars_y','date']].sort_values('user_id')
index_list = list(duplicated.index)
duplicated_avg = duplicated.groupby(['user_id','name']).mean()
duplicated_avg.tail(25)
```

```python
# 중복으로 들어간 (== 유저가 여러번 방문한) 레스토랑에 대해 평균 평점이 4.5 이상인 레스토랑 저장 (이후 가중치 주기 위해서)

double_restaurant = duplicated_avg[duplicated_avg['stars_y'] >= 4.5].reset_index().drop_duplicates(['name'], keep='first')[['name','stars_y']]
```

```python
duplicated_avg.shape
```

```python
one_dupli = duplicated[duplicated.duplicated(['user_id','name'], keep='first')]
```

```python
one_dupli.shape
```

```python
twi_dupli = one_dupli[one_dupli.duplicated(['user_id','name'], keep='first')]
```

```python
twi_dupli.shape
```

```python
twi_index = list(twi_dupli.index)
```

```python
duplied_clear = one_dupli.drop(twi_index, axis=0)
duplied_clear
```

```python
import numpy as np
avg_rating = list(duplicated_avg.stars_y.values)
avg_rating = list(np.round(avg_rating,1))
```

```python
duplied_clear['stars_y'] = avg_rating
duplied_clear
```

```python
not_duplicated = main.drop(index_list, axis=0)
not_duplicated = not_duplicated[['user_id','name','stars_y','date']]
not_duplicated
# duplicated된 df index 기준 빼주고 다시 합치기.
```

```python
matrix_df = pd.concat([not_duplicated,duplied_clear], axis=0)
matrix_df
```

```python
matrix_df[matrix_df.duplicated(['user_id','name'])] # 중복 없음 확인
```

```python
main_index = list(matrix_df.index)
```

```python
final_main = main.loc[main_index]
```
